# 콜금리 crawl

```python
import requests
from bs4 import BeautifulSoup
import os
import pandas as pd
import numpy as np
import time
from datetime import timedelta
```

```python
# 520p 까지
callrate_list = []

for i in range(1, 520):
    print(i,'번 째 페이지를 가져오고 있습니다...')
    url = 'https://finance.naver.com/marketindex/interestDailyQuote.nhn?marketindexCd=IRR_CALL&page={}'.format(i)

    response = requests.get(url)
    # response # response.url # response.content
    soup = BeautifulSoup(response.content, 'html.parser')
    # soup
    content = soup.find('table', class_='tbl_exchange today').find_all('tr')
    # content

    for td in content:
        temp_dict = {}

        date = td.find('td', class_='date')
        if date is not None:
            date = date.text.replace('\n','').replace('\t','')
            temp_dict['date'] = date
            # print(date)

        callrate = td.find('td', class_='num')
        if callrate is not None:
            callrate = callrate.text
            temp_dict['callrate'] = float(callrate)
            # print(callrate)
            
            callrate_list.append(temp_dict)

print(callrate_list)  # len 3633개
```

```python
df = pd.DataFrame(callrate_list)
df['date'] = pd.to_datetime(df['date'])
df = df.set_index(df['date'])
df['old_date'] = None
df['old_callrate'] = None
df['old_date'] = df['date'] - timedelta(days = 30)
df['old_date2'] = df['date'] - timedelta(days = 31)
df['old_date3'] = df['date'] - timedelta(days = 35)
df['old_date4'] = df['date'] - timedelta(days = 40)
df['old_date5'] = df['date'] - timedelta(days = 45)
df['old_date6'] = df['date'] - timedelta(days = 50)
df['old_date7'] = df['date'] - timedelta(days = 55)
df['updown'] = None


# for i in range(len(callrate_list)): # 전체 loop용
for i in range(30): # test용
    print(i,'번 째 루프를 돌고 있습니다...')
    if df['old_date'][i] in list(df['date']):
        df['old_callrate'][i] = df.loc[df['old_date'][i]]['callrate']
        
        while df['old_date'][i] not in list(df['date']):
            df['old_date'][i] = df['date'] - timedelta(days = 1)
            if df['old_date'][i] in list(df['date']):
                df['old_callrate'][i] = df.loc[df['old_date'][i]]['callrate']

    try:
        if (df['callrate'][i] - df['old_callrate'][i]) < 0:
            df['updown'][i] = 'down'
        elif (df['callrate'][i] - df['old_callrate'][i]) == 0:
            df['updown'][i] = 'neutrality'
        else:
            df['updown'][i] = 'up'
    except:
        pass
        
df = df.drop(['old_date3'], axis = 1)
df = df.drop(['old_date4'], axis = 1)
df = df.drop(['old_date5'], axis = 1)
df = df.drop(['old_date6'], axis = 1)
df = df.drop(['old_date7'], axis = 1)
    
print(df)
```

```python
df.to_csv("callrate_ver2.csv")
# 이후 excel을 열어서 순서를 조정해 줄 것
```
