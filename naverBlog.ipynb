```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint
import numpy as np
import pandas as pd
import datetime

news_list = []
NAVER_SEARCH_URL = "https://search.naver.com/search.naver"
query = input(str("키워드를 입력하세요: "))
params = {
            'where': 'post',
            'query': query,
            'start': int(1)
            }


resp = requests.get(NAVER_SEARCH_URL, params=params)
soup = BeautifulSoup(resp.content, 'html.parser')
ul_contents = soup.find('ul', class_='type01')

print(ul_contents)
```

```python
len(ul_contents)
```

```python
# ul_contents.contents를 확인해서, 리스트에 이상이 있는지를 확인합니다.
# 확인해보면, 리스트 안에 빈 리스트가 있다는 것을 확인할 수 있습니다.
new_ul = []
for content in ul_contents:
    ul_contents.contents
    if not str(content).strip():
        continue
    new_ul.append(content)
```

```python
len(new_ul)
```

```python
blogList = []

for n in new_ul:
    blogDict = {}

    a_tag = n.find('dl').find('dt').find('a')
    blogDict['link'] = a_tag['href']
    blogDict['title'] = a_tag.text

    b_tag = n.find('dl').find('dd', class_ = "txt_inline")
    blogDict['timeline'] = b_tag.text

    c_tag = b_tag.find_next('dd')
    blogDict['script'] = c_tag.text

    innerBaseUrl = 'https://blog.naver.com/'
    innerUrl = a_tag['href']
    innercontents = requests.get(innerUrl)
    innersoup = BeautifulSoup(innercontents.content, 'html.parser')
    
    # iframe 
    innerRealUrl = innersoup.find('iframe')['src']
    innerRealContent = requests.get(innerBaseUrl + innerRealUrl)
    innerRealsoup = BeautifulSoup(innerRealContent.content, 'html.parser')
    print(innerRealsoup.find('div', class_='se-main-container').text.replace('\n',''))
    
    
    blogList.append(blogDict)
    break
    
# print(blogList)
    
```
