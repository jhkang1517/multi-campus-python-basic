```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint
```

```python
gaon_chart_url = ('http://www.gaonchart.co.kr/main/section/chart/online.gaon')

params = {
    'nationGbn': 'T',
    'serviceGbn': 'ALL'
}
```

```python
resp = requests.get(gaon_chart_url, params=params)
# resp = requests.get(melon_chart_url)
resp
```

```python
soup = BeautifulSoup(resp.content, 'html.parser')
soup
```

```python
gaon_contents = soup.find('table')
new_ul = []
for content in gaon_contents:
    gaon_contents.contents
    if not str(content).strip():
        continue
    new_ul.append(content)
new_ul
```
