# Action Chain

```python
from selenium.webdriver.common.action_chains import ActionChains
```

```python
driver_path = 'C:\\Users\\student\\Downloads\\chromedriver_win32\\chromedriver'
driver = webdriver.Chrome(driver_path)
driver.implicitly_wait(30)
driver
```

```python
driver.get('https://watcha.com/ko-KR/')
search_elem = driver.find_element_by_xpath('//*[@id="search_bar_in_home"]')

search_elem.send_keys('기생충')
search_elem.send_keys(u'\ue007')

a_link = driver.find_element_by_xpath('//*[@id="root"]/div/div[1]/section/section/section[1]/div[2]/div[1]/div/div/div/ul/li[1]/a')
a_link
```

```python
signup_elem = driver.find_element_by_xpath('//*[@id="root"]/div/div[1]/div/nav/ul/li[2]/button')
signup_elem.click()
```

```python
from selenium.webdriver.common.action_chains import ActionChains

actions = ActionChains(driver)
actions.move_to_element(signup_elem) # 회원가입으로 이동.
actions.click(signup_elem) # 회원가입 클릭
actions.perform()
actions.reset_actions()
driver.implicitly_wait(30)
```

```python
name_input = driver.find_element_by_xpath('//*[@id="sign_up_name"]')
email_input = driver.find_element_by_xpath('//*[@id="sign_up_email"]')
password_input = driver.find_element_by_xpath('//*[@id="sign_up_password"]')
signup_btn = driver.find_element_by_xpath('//*[@id="root"]/div/div[4]/div/div/div/section/div/div/form/button[2]')
driver.implicitly_wait(30)
# action Start
```
