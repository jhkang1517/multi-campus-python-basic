# 여러 그래프 그리기
수식1.
$$
y_1=2x^2   \\
y_2=3x+20  \\
y_3=4x^3-20  \\
$$

```python
x = np.arange(-5, 5, .1)
y1 = 2*x**2
y2 = 3*x+100
y3 = x**3-20
y4 = x**4+10
```

```python
# 방법1. 여러번 호출
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.show()
```

```python
# 방법2. 한번에 호출
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

# 여러 그래프창에 그리기

```python
plt.plot(x, y1) # 처음 호출시 새로운 그래프 창 생성 -> 그래프 그림
plt.figure() # 새로운 그래프 창 생성
plt.plot(x, y2) #새로운 그래프 창에 그림

# plt.figure(n) # qt 일때 확인 가능
```

## clf - 그래프 지우기 (clear figure)

```python
x = np.arange(-5, 5, .1)
y1 = x**2-2
y2 = 20*np.cos(x)**2

plt.figure(1)
# figure 1 창 생성
plt.plot(x, y1)
# 그림그리기

plt.figure(2)
# figure 2 창 생성
plt.plot(x,y2)
# 그리기

plt.figure(1)
# 다시 figure 1 창 호출
plt.plot(x,y2)
# 그리기

plt.figure(2)
# 다시 figure 2 창 호출
plt.clf()
# 비우기
plt.plot(x,y1)
# 다시 그리기

plt.show()
```

# 하나의 그래프 창에서 영역 나누기

```python
# plt.subplot(m, n, p)  mxn 행렬의 p번째 그려짐
plt.subplot()
```

y_1=0.3(x-5)^2 +1 \ y_2=-1.5x+3 \ y_3=sin(x)^2 \ y_4=10e^{-x}+1 $$

```python
x = np.arange(-5, 5, .1)
y1 = 0.3*(x-5)**2 + 1
y2 = -1.5*x + 3
y3 = np.sin(x)**2
y4 = 10*np.exp(-x) + 1

# 2 x 2 행렬로 나누고 위치조정
plt.subplot(2, 2, 1) # 2 by 2로 나눈 뒤, 첫 번째 창에다가 
plt.plot(x, y1) # 그래프 그리기

plt.subplot(2,2,2) # 2 by 2 로 나눈 두 번째 창에
plt.plot(x, y2) # 그래프 그리기

plt.subplot(2,2,3) # 2 by 2 로 나눈 세 번째 창에
plt.plot(x, y3) # 그래프 그리기

plt.subplot(2,2,4) # 2 by 2 로 나눈 4 번째 창에
plt.plot(x, y4) # 그래프 그리기

plt.show()
```

```python
apple_price = [93.95, 112.15, 104.5, 144.85, 169]
ms_price= [39, 50, 57, 69, 95]
year = [2014, 2015, 2016, 2017, 2018]
```

```python
plt.bar(year, apple_price)
```

```python
plt.bar(year,apple_price)
plt.bar(year,ms_price)
plt.show()
```

```python
plt.bar(year,apple_price, width=.5)
plt.bar(year, ms_price, width=.5)
plt.show()
```

```python
year = np.arange(2014,2019)
_year = np.arange(5)

width = .35

plt.bar(_year, apple_price, width=width, label='apple')
plt.bar(_year+width, ms_price, width=width, label='ms')
plt.xticks(_year+width/2, year) # 정확하게 x의 값의 위치를 옮겨준다. 
plt.legend()
plt.show()
```

```python
# 산포도
plt.scatter(year, apple_price)
plt.scatter(year, ms_price)
plt.xlabel('Year')
plt.ylabel('apple_price(apple)')
plt.show()
```

```python
plt.scatter(year, apple_price, s=np.array(apple_price)**3/10000) # 크기에 대한 가중치를 준다
plt.scatter(year, ms_price, s=np.array(ms_price)**3/1000) # 크기에 대한 가중치를 준다
plt.xlabel('Year')
plt.ylabel('apple_price(apple)')
plt.show()
# 사실 y 값이 좌측에 주어지기 때문에 큰 의미는 x 
```

```python
data = {
    'ms_price': np.array(ms_price),
    'apple_price': np.array(apple_price),
}
plt.scatter(year, apple_price, s=data['apple_price']**3/10000)
plt.scatter(year, ms_price, s=data['ms_price']**3/1000)
plt.xlabel('Year')
plt.ylabel('apple_price(apple)')
plt.show()
```

```python

data = {
    'ms_price': np.array(ms_price),
    'apple_price': np.array(apple_price),
}
plt.scatter(year, apple_price, s=data['apple_price']**3/10000)
plt.scatter(year, ms_price, s=data['ms_price']**3/1000)
plt.xlabel('Year')
plt.ylabel('apple_price(apple)')

plt.axis([2013, 2019, 0, 200])
plt.show()
```

```python

# 3d
from mpl_toolkits.mplot3d import Axes3D
# 3d를 import 
import numpy as np
from matplotlib import pyplot as plt

# 3D Sample data만들기
mu_vec1 = np.array([0,0,0])
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]]) 

class1_sample = np.random.multivariate_normal(mu_vec1, cov_mat1, 20)
class2_sample = np.random.multivariate_normal(mu_vec1 + 1, cov_mat1, 20)
class3_sample = np.random.multivariate_normal(mu_vec1 + 2, cov_mat1, 20)

fig = plt.figure(figsize=(8,8)) # figure -> Canvas
ax = fig.add_subplot(111, projection='3d')
   
ax.scatter(class1_sample[:,0], class1_sample[:,1], class1_sample[:,2], 
           marker='x', color='blue', s=40, label='class 1')
ax.scatter(class2_sample[:,0], class2_sample[:,1], class2_sample[:,2], 
           marker='o', color='green', s=40, label='class 2')
ax.scatter(class3_sample[:,0], class3_sample[:,1], class3_sample[:,2], 
           marker='^', color='red', s=40, label='class 3')

ax.set_xlabel('variable X')
ax.set_ylabel('variable Y')
ax.set_zlabel('variable Z')

plt.title('3D Scatter Plot')
     
```

```python
# Histogram
import numpy as np
import random
import math
from matplotlib import pyplot as plt

data = np.random.normal(0, 20, 1000) 
# 1000개 샘플링

bins = np.linspace(math.ceil(min(data)), 
                   math.floor(max(data)),
                   20) # fixed number of bins

plt.xlim([min(data)-5, max(data)+5])
# 최저 -5 부터 최고 +5 까지

plt.hist(data, bins=bins, alpha=0.5)
# alpha 값은 밝기를 의미합니다. 
plt.title('Gaussian')
plt.xlabel('variable X (20 evenly spaced bins)')
plt.ylabel('count')

plt.show()
```
