```python
import requests
```

```python
# HTTPS (통신 규약)
# requests.request(method, url, **kwargs)
# requests.<http_method>(url, params=None(세부사항), **kwargs)
# resp(변수)
resp = requests.request('get', 'https://naver.com')
```

```python
resp
# 클래스로 취급
# 응답 코드가 나타난다.
# [200] = 응답코드. 2xx: 성공
# 3xx = redirection 리 다이렉션. 다른쪽으로 이동되었다. 그곳으로 안내하겠다. 
# 4xx = 요청 오류. 
# 5xx = 서버 오류.
```

```python
resp = requests.get('https://search.naver.com/search.naver', params={'query': '무역전쟁'})
```

```python
resp
```

```python
# unicode text
resp.text
# 이 안의 메세지는 뭔가요?
# 유니코드 방식
```

```python
# byte content
resp.content
# 맨 처음 b가 붙는다.
# 문자열이 바뀐다.
# 메세지의 body 부분을 가져온다. 
```

```python
resp.encoding
# UTF-8 의 형식이다. 
```

```python
resp.status_code
# response에 대한 응답 코드 
```

```python
resp.status_code == requests.codes.ok
```

```python
requests.codes.ok
# 변수입니다. = 200
```

```python
if resp.status_code != requests.codes.ok:
    """
    """
    pass
```

```python
resp.headers
# 데이터 요청 자체에 대한 header
```

```python
resp.url
# 요청에 대한 URL
```

```python
resp.json()
# requests의 형태가 json일 경우 바로 반환할 수 있다. 
# 지금은 html의 형식이기 때문에 오류가 발생한다. 
```

## JSON실습

```python
json_url = "https://crix-api-cdn.upbit.com/v1/crix/candles/days?code=CRIX.UPBIT.KRW-BTC&count=200"

resp2 = requests.get(json_url)
```

```python
json_content = resp2.json()
json_content
```

```python
type(json_content)
```

```python
type(json_content[0])
# 리스트의 0번째 순서 = 딕셔너리 
```

```python
for json in json_content:
    print(json)
```

```python
print(json)
# json은 리스트 형태이기 때문에, 하나하나 꺼내기 위해서는 for문을 사용해야 합니다.
```
