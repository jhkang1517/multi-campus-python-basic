```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint
import numpy as np
import pandas as pd
import datetime
```

```python
news_list = []

NAVER_SEARCH_URL = "https://search.naver.com/search.naver"
query = input(str("키워드를 입력하세요: "))

# for문을 돌립니다. 왜 why? 한페이지만 가져올 것이라면 필요없지만, 여러개의 페이지에 접근할 것이기 때문입니다.
# params의 형식은 꼭 사용하는 것은 아니지만, 요즘 대세입니다. 
# range(0, 9) << 9페이지 까지 가져옵니다. 9의 숫자를 바꾸면 원하는 만큼 범위가 늘어납니다.
for i in range(0,5):
    params = {
        'where': 'post',
        'query': query,
        'start': int((i*10)+1)
        # int((1*10)+1)로 설정합니다. 
        #왜 why? 네이버 블로그 페이지에는 10개가 출력되기 때문에, 다음페이지에는 11이 필요하기 때문입니다.
    }
    
    # response와 soup를 만듭니다. 
    # ul_contents의 경우, 원하는 부분만 확인하면 되기 때문에, 홈페이지에서 범위를 명확하게 정해줍니다. 
    # ul_contents.contents 함수를 사용하여 정상적인지 확인해 봅니다.
    resp = requests.get(NAVER_SEARCH_URL, params=params)
    soup = BeautifulSoup(resp.content, 'html.parser')
    ul_contents = soup.find('ul', class_='type01')
    
    
    # ul_contents.contents를 확인해서, 리스트에 이상이 있는지를 확인합니다.
    # 확인해보면, 리스트 안에 빈 리스트가 있다는 것을 확인할 수 있습니다.
    new_ul = []
    for content in ul_contents:
        ul_contents.contents
        if not str(content).strip():
            continue
        new_ul.append(content)
    
    # 딕셔너리를 만들어줍니다. 
    # for문을 돌립니다. 왜 why? 딕셔너리를 계속 초기화 시켜주면서 리스트에 추가할 것이기 때문입니다. 
    for li in new_ul:
        news_dict = {}
        
        a_tag = li.find('dl').find('dt').find('a')
        news_dict['link'] = a_tag['href']
        news_dict['title'] = a_tag.text

        b_tag = li.find('dl').find('dd', class_ = "txt_inline")
        news_dict['timeline'] = b_tag.text
        
        c_tag = b_tag.find_next('dd')
        news_dict['script'] = c_tag.text
        
        news_list.append(news_dict)
    
print(datetime.datetime.now(),"기준")
print(int(len(news_list)), "개의 결과값을 가져왔습니다.")
pprint(news_list)
```
