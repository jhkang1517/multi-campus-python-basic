# Abstract

#### 연구 배경
점점 더 많은 고객들이 편리성, 안정성 및 저렴한 비용을 지닌 온라인 쇼핑을 선택하고 있다.  
온라인에서 판매되는 물품들 역시 증가함에 따라, 단순한 사진이나 짧은 상품소개로는 고객들의 구매를 유도할 수 없다.  
반면에, 텍스트 기반의 고객리뷰는 상품들의 비교, 경험들이 매우 풍부하고, 이는 상품을 구매하도록 하는데에 밑거름이 된다.   
온라인 쇼핑몰인 아마존의 경우, 고객들이 리뷰를 남기는 것을 허락한다.  
그들의 리뷰는 다른 고객들에게 믿을 수 있는 자원(source)가 된다.   
전통적으로, 많은 고객들은 상품수가 제한된 전문적인 순위를 참고했다.  
이러한 방식은 일반적인 제품의 품질에 따라 순위를 매겼습니다.  
그러나 일반적으로 제품들은 여러 특징들을 가지고 있고, 다른 역할들을 수행한다.   
따라서 고객들은 단순한 품질이 아닌 자신의 취향, 생활환경에 맞는 여러 기능들에 각각 관심이 있을 수 있습니다.  
 
#### 구현 방향
따라서, 이 논문에서 연구자들은 feature-based product에 ranking을 매겨볼 것입니다.  
이를 위해서 수천개의 customer's reviews를 사용할 것입니다.   
 
 * 첫번째로, 먼저 제품 카테고리 내의 제품 기능을 확인하고 빈도 및 상대적 사용을 분석합니다. 각각의 특징에 따라 우리는 리뷰속의 문장들로 이들을 비교해 볼 것입니다.
 
 * 그 다음, 각 문장마다 감성의 방향이 어느쪽인지를 파악하고, (점수를 매깁니다?) 고객 리뷰에서 얻은 정보를 사용하여 가중치 및 방향 그래프를 구성하여 제품 간 관계를 모델링합니다.그런 다음 제품의 상대적 품질을 결정하기 위해 그래프를 분석합니다. Amazon.com의 실제 데이터에서 디지털 카메라 및 TV 리뷰 데이터를 사용하여 연구자들이 제안한 분석방법의 결과를 나타낼 것입니다.

## 1. Introduction
#### 연구배경
  인터넷 연결이 급속하게 확산되면서 전자 상거래의 규모는 점점 커져가고 있다.    
  소비자들 역시 편리하고 안정적이고 빠른(!) 온라인 쇼핑으로 전환하고 있다.    
      
#### 이 연구 왜했어? 왜 Why?!  
  판매자가 제공한 이미지, 제품 설명만을 근거로 구매 결정을 내리는 것은 이제 너무 어려움   
  온라인 판매 업체 역시 가격, 제조 업체 및 기타 특성과 같은 특정 요구에 따라 제품을 선택할 수 있게 하여 소비자에게 포괄적인 쇼핑 경험을 제공하는 것을 목표로 하고 있다.  
      
#### 그러면 어떻게 해야하는데?!   
  고객은 '구입한 제품에 대해서' (question1. 구매하지 않으면 리뷰를 할 수 없는가?) 리뷰를 남길 수 있다.  
  이러한 고객 리뷰는 일반적으로 판매자들이 제공한 설명보다 더 정직하고, 편파적(?)이며 포괄적(?)인 것으로 간주된다.  
  특히 비슷한 요구사항을 가진 고객들은 그와 비슷한 고객들이 남긴 리뷰에게 많은 영향을 받는다.  
    
#### 그 다음 문제는?!    
  문제는 고객들의 리뷰 역시 기하급수적으로 많다는 것이 가장 큰 문제    
  따라서 유용한 정보를 추출하기 위해서는 이 리뷰들을 효율적으로 마이닝해야한다.  
  
#### And How?  
  "Our work is based on the our proposed ranking scheme, where products are ranked based on their overall quality."    
  연구자들의 작업은 기본적으로는 제품의 전반적인 품질에 따라 매겨진 순위에 기반한다.    
  그러나, 단순한 전반적인 품질에 매겨진 순위와 더불어, 제품들마다 상이한 기능 역시 중요하다.  
  예를 들어, 전반적으로 평점이 높더라도, 배터리의 수명이 적다면 오래가는 배터리를 원하는 소비자들은 만족하지 않을 것이다.   
    
  따라서, 전반적인 순위보다 상세한 제품들의 기능이 고려된 순위가 중요하다.  
  이 논문에서 연구자들은 "고객 리뷰 텍스트(customer review text)"를 사용할 것이고, 수만 건의 리뷰를 통해 제품의 순위를 지정하는 알고리즘을 제안할 것이다.  
    
  각 제품의 범주에 대한 "제품 기능(product feature)"이 데이터 전처리 과정에서 정의되고 추출될 것이다.  
  대부분의 제품에는 중요한 것으로 간주되는 표준 기능 세트가 존재하므로 이를 사용한다.  
    
  그 다음, 각 문장에 설명된 제품 기능을 labeling한다.  
  그 다음, 제품 평가에서 유용한 4개의 유형의 고객 리뷰 문장을 확인한다.  
  긍정적이면서 주관적인, 부정적이면서 주관적인, 긍정적인 비교, 부정적인 비교가 이 4개의 유형입니다.    
  주관적인 문장은 리뷰어가 하나 이상의 제품 기능의 긍정적 또는 중요성을 표현하는 문장입니다.  
  

## 3. Product Ranking Methodology 연구 방법
### 3.1 Identifying Product Features 제품 기능 식별
 - 제품 기능은 제품에 추가 기능을 제공하는 속성 
     - 제조업체는 서로 다른 관점에서 제품 기능을 강조하곤 함
     - 이러한 제품 기능들의 조합은 고객의 구매결정에 영향을 미침 
     - 몇몇 소비자들은 다른 소비자들과는 다르게 여러 기능의 조합을 고려함

 - 구현 방법
     - 공식적인 소비자 보고서를 기반으로 디지털 카메라 및 텔레비전의 두 카테고리에 대한 제품 기능 세트를 수집
     - 같은 제품 기능이지만 동의어를 가지기도 하기 때문에, 직접 수동으로 레이블을 지정해 줌
     - 총 문장수 = 1,516,001 개 
     - 이 데이터의 집합에서 16%의 문장이 우리가 수집한 특징 중 하나 이상을 기술한다는 것을 확인함
 
 - 각 문장에 특징은 어떻게 붙일까?
     - 문장에 제품 기능에 대한 단어/구문 중 하나라도 있는 경우, 해당 특징을 설명하는 것으로 표시

#### 디지털 카메라의 경우
   - 실내 및 저조 립 사진 촬영을 위해 플래시가 중요한 기능입니다.
   - 배터리 수명은 사용 된 배터리의 종류를 자세히 알려주는 기능입니다.
   - 초점은 자동 초점 또는 수동 초점 기능에 대해 설명합니다.
   - 렌즈는 고급 카메라를 구입하는 전문 사진 작가에게 중요한 요소입니다.
   - 광학 기능은 디지털 줌과 광학 줌을 포함합니다. 
   - LCD는 사용자가 사진이 어떻게 보일지를 볼 수있는 디지털 디스플레이 / 스크린을 나타냅니다.
   - 해상도는 그림의 선명도 또는 세부 사항을 나타냅니다. 연속 촬영은 급속 사격 및 연속 사격 기능을 설명하는 데 사용됩니다.
   - 메모리는 촬영할 수있는 사진의 수를 결정하고 압축은 사진의 파일 크기가 축소되는 방법을 결정합니다. 
     
#### 텔레비전 세그먼트의 경우
   - 사운드 기능은 오디오 품질에 관심이있는 사용자에게 유용합니다 (일부 TV에는 서라운드 사운드를 생성하기 위해 별도의 스피커 세트가 제공됨).
   - 반사 효과 / 반사 방지는 시청 경험에 중요합니다.
   - 크기는 텔레비전 화면의 크기, 높이, 무게를 나타냅니다.
   - 연결은 장치를 TV에 연결하는 데 사용할 수있는 입력 포트의 수와 유형을 의미합니다. 표시된 이미지의 풍부함 / 품질은 화질에 설명되어 있습니다.
   - 사용자는 텔레비전에서 사용 가능한 원격 제어 장치에도 관심이 있습니다.
   - 해상도는 화면에 표시되는 픽셀 수 또는 라인 수를 나타냅니다.
   - 조정은 화면을 채우기 위해 이미지를 확장하거나 압축하는 기능 / 모드입니다.
   - PiP (picture-in-picture) 기능을 통해 사용자는 한 번에 두 개의 채널을 시청할 수 있습니다. 
   - Film-Mode / CineMotion은 일부 사용자에게 중요 할 수있는 영화 감상 경험을 향상시킵니다.
  
### 3.2 Sentence labeling 문장 라벨링
 - 고객은 제품에 대한 의견을 여러가지 방법으로 표현
     - 이 중 중요한 두가지 문장, "주관적 문장","비교 문장"
     - 연구자들은 MxTerminator를 사용하여 리뷰를 문장으로 분리함
     
### 3.3 비교문 식별
 - 키워드 비교를 사용하여 비교문을 식별한다.
     - 명시적인 키워드 ("우수함, 초과, 비교, 우수")
     - 암시적인 키워드 ("선호, 선택 등")

#### 구현 어떻게 해요?!
 - 기존에는 자바기반의 조건부 랜덤 필드 품사 (POS) 태그를 사용함
     - 문장에 KW의 비교 키워드가 포함되어 있는지 확인(?)
     - 모든 단어를 POS 태그로 인식할 것. 
     - 특히 {JJR (비교 형용사), RBR (비교 형), JJS (최상급 형용사), RBS (최상급 부사)}
     - MQPA 사전에 정의된 구조 패턴(wordset)이 문장에 존재하는지 검색한다. (주관적, 비교 둘 다)
     - 그러나, 반드시 비교문장이 제품을 비교하는 것은 아닐 수 있다.
     
#### 따라서 연구자들은 위를 진행 한 후, 새로운 방법을 사용
 - 비교문에서 제품과 제품을 비교하는 쌍을 식별하기 위해 동적 프로그래밍 기술을 사용
     - 순위 모델을 작성하는 동안 문장이 설명하는 제품과 다른 적어도 하나의 제품 이름을 포함하는 비교 문장만을 사용
     - 82%의 정확도와 80%의 리콜이 나타남
     
## 3.4 문장 감정 확인 
 - 문장이 어떠한 정서(감성)을 지니고 있는지를 확인
     - 긍정적이고 부정적인 정서만 고려
     - 안타깝게도 WordNet과 같은 사전이나 다른 출처에는 각 단어에 대한 정서 지향 정보가 포함되어 있지 않다.
     - 이 논문에서는 MQPA 프로젝트에서 개발 된 포지티브 워드 세트 (POS)와 네가티브 워드 세트 (NEG)를 활용하여 간단하면서도 강력한 방법을 사용
     - 연구자들은 또한 연구자들이 고려한(?) 몇 가지 단어를 추가합니다.
     - 이 과정이 끝나면 양성 세트에 대한 1974 단어와 음화 세트에 대한 4605 단어의 목록을 얻는다.
     - 문장에 긍정적/부정적인 단어 집합에 포함 된 단어가 포함되어 있으면 긍정적/부정적 인 태그를 붙임
     - 부정적인 감정 단어도 비슷하게 처리됩니다.
     
     - 그러나 많은 고객들은 독단적 인 문장을 사용하면서 자신의 의견을 표현하는 것을 싫어하지만 리뷰에 약간의 부정을 사용한다. 따라서
     - 우리는 28 개의 부정 (negation) 4 단어를 수동으로 만듦
     - 감정 표현의 동일한 목록을 사용하여 비교 문장의 문장 방향을 결정한다는 것을 언급해야만 한다.
     - 82 %의 정확도와 대략 86 %의 리콜을 달
