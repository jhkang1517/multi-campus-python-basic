```python
# nltk 라이브러리와 brown corpus import
import nltk
from nltk.corpus import brown
```

```python
# brown download
nltk.download('brown')
```

```python
# 말뭉치의 모든 장르 확인
print(brown.categories())
```

```python
# 3가지 genres를 선택하고 7가지 whwords 생성
genres = ['fiction','humor','romance']
whwords = ['what', 'which', 'how', 'why', 'when', 'where', 'who']
```

```python
# 세고 싶은 단어를 세는 for 구문 
for i in range(0, len(genres)):
    genre = genres[i]
    print()
    print("'" + genre + "' wh 단어 분석 ")
    genre_text = brown.words(categories = genre)
    # nltk 라이브러리의 FreqDist 객체를 사용하는 구문 복합 
    fdist = nltk.FreqDist(genre_text)
    for wh in whwords:
        print(wh + ':', fdist[wh], end=' ')
```

## 웹 및 채팅 텍스트 자료 파일 중 하나에서 빈도 분포 작업 탐색

```python
import nltk, matplotlib
nltk.download('webtext')
from nltk.corpus import webtext
print(webtext.fileids())
```

```python
# 개인 광고 데이터가 포함된 파일을 선택하고 빈도 분포를 확인
fileid = 'singles.txt' # 개인 광고 데이터
wbt_words = webtext.words(fileid)
fdist = nltk.FreqDist(wbt_words)
# fdist.max()
print('최대 발생 토큰 "', fdist.max(), '" 수 : ', fdist[fdist.max()])
# fdist.N()
print('말뭉치 내 총 고유 토큰 수 : ', fdist.N())
# 빈번하게 쓰인 단어 
print('말뭉치에서 가장 흔한 10개 단어는 다음과 같습니다.')
print(fdist.most_common(10))
# fdist.tabulate() 함수로 표로 만들기
print('개인 광고의 빈도 분포')
print(fdist.tabulate())
```

```python
# 누적 빈도가 있는 도수 분포 그래프
fdist.plot(cumulative=True)
```
