# 외부 말뭉치 다운로드, 로드하고 엑세스하기

```python
from nltk.corpus import CategorizedPlaintextCorpusReader
```

```python
# C:\Users\student\Documents\python_basic\python_NLP_cookbook
reader = CategorizedPlaintextCorpusReader(r'/Users/student/Documents/python_basic/python_NLP_cookbook/mix20_rand700_tokens_cleaned/tokens', r'.*\.txt', cat_pattern=r'(\w+)/*')
# r' 의 의미는?
# row텍스트의 의미로, 입력한 대로 사용한다는 의미이다. 


print(reader.categories())
print(reader.fileids())
```

```python
posFiles = reader.fileids(categories='pos')
negFiles = reader.fileids(categories='neg')
```

```python
from random import randint
fileP = posFiles[randint(0, len(posFiles)-1)]
fileN = negFiles[randint(0,  len(negFiles) - 1)]
# randint(인자가 두 개 필요)

print(fileP)
print(fileN)
```

```python
for w in reader.words(fileP):
    print(w + ' ', end='')
    if (w is '.'):
        print()
for w in reader.words(fileN):
    print(w + ' ', end='')
    if (w is '.'):
        print()
```
