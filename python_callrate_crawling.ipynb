```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
import numpy as np
import time
from datetime import datetime

call_list = []

for i in range(1, 445):

    url = 'https://finance.naver.com/marketindex/interestDailyQuote.nhn?marketindexCd=IRR_CALL&page={}'.format(i)

    response = requests.get(url)
    soup = BeautifulSoup(response.content)
    content = soup.find('table', class_='tbl_exchange today').find_all('tr')

    for tag in content:
        call_dict = {}

        # date 
        date = tag.find('td')
        if date is not None:
            date = date.text.replace('\n','').replace('\t','')
            call_dict['date'] = date
            print(date)

        # price
        if tag.find('td') is not None:
            price = tag.find('td').find_next('td').text
            call_dict['price'] = price

            call_list.append(call_dict)
```

```python
df = pd.DataFrame(call_list)
df = df.set_index('date')
df
```
