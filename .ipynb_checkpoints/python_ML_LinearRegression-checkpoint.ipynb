# LinearRegression_model_comprehension

```python
from sklearn.datasets import make_regression # 가상의 회귀분석 데이터 생성
from sklearn.linear_model import LinearRegression # LinearRegression 모델 객체
from sklearn.model_selection import train_test_split # 나누는 함수 
import numpy as np
```

```python
X, y = make_regression(n_samples=1000, n_features=1, bias=10, noise=10)
print('input_feature:\n {}\nTarget_value: \n{} '.format(X[:5], y[:5]))

# sample = 전체 데이터 개수
# features = 전체 변수의 개수 
# bias = 편향. 절편
# noise = 소음. 이상치..? (확인필요)
```

```python
print(type(X), type(y)) # numpy의 배열 
```

```python
%matplotlib inline 
# mat + plot + libreary 라이브러리 
import matplotlib.pyplot as plt

plt.scatter(X[:, 2], y) # 산포도 # 전체행렬에서 1열 데이터를 찝는다.
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
# train_test_split을 통해서 사이즈 나누기
# shuffle: 섞기 
```

```python
X_train # 8000개 (위에서 8:2로 나눔)
```

```python
# Linear Regression 클래스 객체 생성 
model = LinearRegression() # fit_intercept = 상수항 유무 결정 

# fit 메서드로 모형 추정 (오그멘테이션은 자동으로 수행) 
model_lr = model.fit(X_train, y_train) # lr? : linear regression

print('weight: ', model_lr.coef_) # 기울기 
print('bias: ', model_lr.intercept_) # 절편

# weight는 10개가 나오는데, 위에서 feature를 10개로 설정했기 때문이다.
```

#### feature 1000, n_samples 1000 으로 설정하면 과적합 overfitting 발생 
#### feature 수가 너무 높으면 제대로 작동하지 않는다. 즉 PCA, 주성분 분석을 함으로써 feature수를 줄여주어야한다.

```python
model.score(X_train, y_train) # 100을 곱해서 퍼센트로 환산. noise를 올릴수록 정확도가 낮아진다. 
```

```python
model.score(X_test, y_test) # 100을 곱해서 퍼센트로 환산. noise를 올릴수록 정확도가 낮아진다. 
```

```python
model.predict(np.array([1,2]).reshape(2,1))

# fir과 predict의 차이 
```

# RidgeRegression_model_comprehension

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
```

##### 데이터생성

```python
X, y = make_regression(n_samples=10000, n_features=1000, bias=10, noise=100)
# 일부러 과적합이 발생할 것으로 예상되는 모델을 생성 

X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True)
```

```python
# Linear Regression 클래스 객체 생성
model = Ridge(alpha=0.1) 
# fit 메서드로 모형 추정 (오그멘테이션은 자동으로 수행)
model_lr = model.fit(X_train, y_train)

print('weight: ', model_lr.coef_)
print('bias: ', model_lr.intercept_)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test) # Ridge는 정확도는 낮지만, test와 score의 차이가 작다.
```

# LassoRegression_model_comprehension

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
```

```python
X, y = make_regression(n_samples=10000, n_features=100, bias=10, noise=20)

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Linear Regression 클래스 객체 생성
model = Lasso() # fit_intercept = 상수항 유무 결정

# fit 메서드로 모형 추정 (오그멘테이션은 자동으로 수행)
model_lr = model.fit(X_train, y_train)

print('weight: ', model_lr.coef_) # Lasso 의 경우에는 0이 된 값들이 많은 것을 확인할 수 있다. 
print('bias: ', model_lr.intercept_)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
