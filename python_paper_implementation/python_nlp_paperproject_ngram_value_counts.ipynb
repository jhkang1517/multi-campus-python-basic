```python
import pandas as pd 
import numpy as np

nomerge_df = pd.read_csv(' .csv', index_col=0)
```

```python
nomerge_df['ngrams'] = ""
nomerge_df['ngrams'] = nomerge_df['tokens'] + nomerge_df['ngram']
print(len(nomerge_df['ngrams'][1]))
print(len(nomerge_df['tokens'][1]))
print(len(nomerge_df['ngram'][1]))
print(len(nomerge_df['ngram'][1]) + len(nomerge_df['tokens'][1]))
```

```python
nomerge_df.drop(['title','firm','report','tokens','ngram','old_date','callrate','old_callrate'], axis='columns', inplace=True)
```

```python
nomerge_df = nomerge_df.rename(columns={'ngrams':'ngram'})
```

```python
nomerge_df.to_csv("(완료)_채권분석보고서_ngram분석용.csv")
```

```python
nomerge_df.head()
```

```python
temp_ls = []
# final_df = pd.DataFrame(columns='ngram')

for i in range(len(nomerge_df)):
    try:
        test = nomerge_df['ngram'][i].split(',')
        temp_ls.extend(test)
    except:
        pass
# print(temp_ls)

temp_df = pd.DataFrame(temp_ls, columns=['ngram'])

final_df = pd.DataFrame(temp_df['ngram'].value_counts())
final_df
```

```python
final_df.to_csv('된거냐.csv')
```
